## **Limpeza e Transformação**

```
#Transforma em dataframe pandas
dataset = pd.read_csv('SEMANAL_MUNICIPIOS-2019.csv')
dataset.head()
```

```
#Verifica os campos disponíveis
dataset.columns
```

```
#Ordenação por cidade e por unidade de medida
df = dataset.sort_values(['MUNICÍPIO', 'UNIDADE DE MEDIDA'])
df.head(10)
```

```
#Padroniza a DATA FINAL para forma entendível
df['DATA FINAL'] = df['DATA FINAL'].astype('datetime64')

#Faz transformações em campos nulos, padroniza os valores decimais e converter os valores para float
valores = ['PREÇO MÉDIO REVENDA', 'DESVIO PADRÃO REVENDA', 'PREÇO MÍNIMO REVENDA', 'PREÇO MÁXIMO REVENDA', 'MARGEM MÉDIA REVENDA', 'COEF DE VARIAÇÃO REVENDA', 'PREÇO MÉDIO DISTRIBUIÇÃO', 'DESVIO PADRÃO DISTRIBUIÇÃO', 'PREÇO MÍNIMO DISTRIBUIÇÃO', 'PREÇO MÁXIMO DISTRIBUIÇÃO', 'COEF DE VARIAÇÃO DISTRIBUIÇÃO']

for iten in valores:
  df[iten] = df[iten].str.replace(',', '.')
  df[iten] = df[iten].str.replace('-', '0')
  df[iten] = df[iten].str.replace(' ', '0')
  df[iten] = df[iten].astype('float')

df
```

```
#Cria o campo MÊS com base no DATA FINAL. Exemplo: 28 - 07 - 2019 : Mês -> (Julho : 07) ou 201907 (Melhor forma considerando que pode ser analisado um período maior que um ano)
df['MÊS'] = df['DATA FINAL'].map(lambda x: 100*x.year + x.month)

df.head()
```

## **Item a) Média de valores por cidade**

Apresenta as médias de valores de cada combustível por cidade, agrupado por mês

```
#Função retorna a buscar por cidade
def mediaCidade(cidade='SAO PAULO', indicador='PREÇO MÉDIO DISTRIBUIÇÃO'):
  
  print('Para a cidade de ', cidade, ' o indicador de ', indicador, 'apresenta os seguintes resultados\n')

  #Verifica cada indicador do município
  medida = df[['MÊS', 'MUNICÍPIO', 'PRODUTO', indicador]]

  #Criar uma tabela padrão somente com os meses em análise
  dt = pd.DataFrame(np.unique(df['MÊS']), columns=['MÊS'])

  city = medida[medida['MUNICÍPIO'] == cidade]

  for produto in np.unique(city['PRODUTO']):

    produtos = city[city['PRODUTO'] == produto]

    t = produtos.groupby(['MÊS']).mean().reset_index()
    table = pd.DataFrame(t)
    table.columns = ['MÊS', produto]
    dt = pd.merge(dt, table)

  return dt
```

```
# Exemplos de cidades: MARINGA, FORTALEZA, RECIFE, PATO BRANCO
# Indicadores : PREÇO MÉDIO REVENDA, DESVIO PADRÃO REVENDA, PREÇO MÍNIMO REVENDA, PREÇO MÁXIMO REVENDA, MARGEM MÉDIA REVENDA, PREÇO MÉDIO DISTRIBUIÇÃO
mediaCidade()
```

```
#Apresentação da variação mensal

x = mediaCidade()
f, ax = plt.subplots(figsize=(18,8))
ax.set_ylabel('Valor')
ax.set_title('Variação média mensal de produtos')
plt.bar(x.iloc[:,0], x.iloc[:,2], label = 'GASOLINA COMUM')
plt.bar(x.iloc[:,0], x.iloc[:,6], label = 'ÓLEO DIESEL S10')
plt.bar(x.iloc[:,0], x.iloc[:,5], label = 'ÓLEO DIESEL')
plt.bar(x.iloc[:,0], x.iloc[:,1], label = 'ETANOL HIDRATADO')
plt.bar(x.iloc[:,0], x.iloc[:,4], label = 'GNV')
plt.legend()
plt.show()

f, ax = plt.subplots(figsize=(18,4))
ax.set_ylabel('Valor')
ax.set_title('Variação média mensal de produtos')
plt.bar(x.iloc[:,0], x.iloc[:,3], label = 'GLP')
plt.legend()
plt.show()
```

## **Item b) Calcule a média de valor do combustível por estado e região**

```
#Função faz o agrupamento de dados por estado
def mediaEstado(estado='SAO PAULO', indicador='PREÇO MÉDIO DISTRIBUIÇÃO'):

  #Verifica cada indicador do município
  medida = df[['MÊS', 'ESTADO', 'PRODUTO', indicador]]

  #Criar uma tabela padrão somente com os meses em análise
  dt = pd.DataFrame(np.unique(df['MÊS']), columns=['MÊS'])

  state = medida[medida['ESTADO'] == estado]

  for produto in np.unique(state['PRODUTO']):

    produtos = state[state['PRODUTO'] == produto]

    t = produtos.groupby(['MÊS']).mean().reset_index()
    table = pd.DataFrame(t)
    table.columns = ['MÊS', produto]
    dt = pd.merge(dt, table)

  return dt
```

```
#Função faz o agrupamento de dados por região
def mediaRegiao(regiao='SUL', indicador='PREÇO MÉDIO DISTRIBUIÇÃO'):

  #Verifica cada indicador do município
  medida = df[['MÊS', 'REGIÃO', 'PRODUTO', indicador]]

  #Criar uma tabela padrão somente com os meses em análise
  dt = pd.DataFrame(np.unique(df['MÊS']), columns=['MÊS'])

  state = medida[medida['REGIÃO'] == regiao]

  for produto in np.unique(state['PRODUTO']):

    produtos = state[state['PRODUTO'] == produto]

    t = produtos.groupby(['MÊS']).mean().reset_index()
    table = pd.DataFrame(t)
    table.columns = ['MÊS', produto]
    dt = pd.merge(dt, table)

  return dt
```

## **Item d) Quais são as 5 cidades que possuem a maior diferença entre o combustível mais barato e o mais caro.**

```
#Cria as diferenças entre o menor e maior valor de combustíveis
df['DIFERENÇA DO PREÇO DISTRIBUIÇÃO'] = df['PREÇO MÁXIMO DISTRIBUIÇÃO'] - df['PREÇO MÍNIMO DISTRIBUIÇÃO']
df['DIFERENÇA DO PREÇO DE REVENDA'] = df['PREÇO MÁXIMO REVENDA'] - df['PREÇO MÍNIMO REVENDA']
df.head(3)
```

```
def difCidades(id=5, medida='DIFERENÇA DO PREÇO DISTRIBUIÇÃO'):
  print('As',id, 'cidades com maiores diferença em',medida, 'são:\n')
  df2 = df.sort_values([medida], ascending=False)
  candidados = df2['MUNICÍPIO'].tolist()
  print(candidados[0])
  classificados = []
  classificados.append(candidados[0])
  ix = 0

  for item in range(len(df2)) : 

    if candidados[item] not in classificados :
      print(candidados[item])
      classificados.append(candidados[item])
      ix += 1
    else :
      if ix == (id-1):
        break;
```

```
#Informa as cidades com maiores diferenças considerando uma semana
difCidades()
```

## **Item c) Calcule a variância e a variação absoluta do máximo, mínimo de cada cidade, mês a mês**

### **VARIÂNCIA**

image.png


CV = Coeficiente
X = Média

e 

Variancia = S²

```
#Crias os campos com variancia para cada amostra
S = df['PREÇO MÉDIO REVENDA'] * df['COEF DE VARIAÇÃO REVENDA']
df['VARIANCIA DE REVENDA'] = S**2

S = df['PREÇO MÉDIO DISTRIBUIÇÃO'] * df['COEF DE VARIAÇÃO DISTRIBUIÇÃO']
df['VARIANCIA DE DISTRIBUIÇÃO'] = S**2

df.head(2)
```

```
def variancaMunicipio(cidade='SAO PAULO') : 
  print('As variância de preço de REVENDA e DISCRIBUIÇÃO para a cidade ',cidade, ' são:\n')
  exemplo = df[df['MUNICÍPIO'] == cidade]
  exemplo

  return exemplo.groupby(['MÊS']).mean().reset_index()
```

```
#Apresenta os resultados de variância por cidade
aux = variancaMunicipio()
pd.DataFrame([aux.iloc[:,0], aux.iloc[:,15], aux.iloc[:,16]]).T
```
