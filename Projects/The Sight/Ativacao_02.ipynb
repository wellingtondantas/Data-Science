```
# Basic setup
import numpy as np
import matplotlib.pyplot as plt
```

```
# Plot the sigmoid function
z = np.linspace(-5, 5, 200)

def sigmoid(z):
  return 1.0/(1.0 + np.exp(-z))

def tanh(z):
  return (np.exp(z) - np.exp(-z))/(np.exp(z) + np.exp(-z))

def relu(z):
  return 1. * (z > 0)

# Plot activations
plt.figure(figsize=(9,7))
plt.plot(z, sigmoid(z), "b-", linewidth=2, label='Sigmoid', alpha=0.8)
plt.plot(z, tanh(z), "r-", linewidth=2, label='tanh', alpha=0.8)
plt.plot(z, relu(z), "g-", linewidth=2, label='ReLU',alpha=0.8)

# Format plot
plt.plot([-5, 5], [1, 1], 'k--')
plt.plot([-5, 5], [0, 0], 'k--')
plt.plot([-5, 5], [-1, -1], 'k--')
plt.axis([-4, 4, -1.2, 1.2])
plt.legend()
plt.grid(True)
plt.title("Activation Function", fontsize=14)
plt.xlabel('WX+b')
plt.ylabel('Activation Output')
plt.show()
```

Aqui são apresentados três tipos de funções de ativação, **sigmoid**, **tangente hiperbólica**, e **degrau**. São muito utilizados em redes neurais.

A função **sigmoide** produz valores no intervalo [0, 1], é um interessante para problemas do tipo binário, a função de ativação calcula estimativas muito altas de um padrão pertencer a classe ou a outra.

A função **tangente hiperbólica** produz valores no intervalo [-1, 1], tem uma função semelhante a sigmoide, porém é considerada mais relevante que a sigmoide porque a sua derivada converge mais rapidamente. Interessante para problemas que utiliza o gradiente descendente.


A função **degrau** produz somente os valores (0 ou 1), pode ser utilizado para problemas lineares, pessoalmente somente utizei essa função de ativação com o percetron simples (uma das primeiras redes neurais).
