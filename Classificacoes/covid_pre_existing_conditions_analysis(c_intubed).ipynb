```
#Biblitecas iniciais
import numpy as np
import pandas as pd
```

```
#Carregamento dos dados
df = pd.read_csv('https://raw.githubusercontent.com/wellingtondantas/Data-Science/master/Datasets/covid_pre_condition.csv')
df
```

```
#Retira as colunas não essenciais
df=df.drop(columns=['id','entry_date','date_symptoms','date_died'],axis=1)
df
```

```
#Transfere as informações para outra base
dfe = df
```

# **Limpeza de dados**

```
#Substituindo 97, 98, 99 para valores nulos e excluindo
#Porém, na idade, realmente uma pessoa pode ter 97,98,99 anos
dfe.iloc[:,5:]=dfe.iloc[:,5:].replace({97:np.nan,98:np.nan,99:np.nan})
dfe=dfe.dropna()
dfe
```

```
dfe1=dfe[dfe['icu']==2][1:3000]
dfe2=dfe[dfe['icu']==1][1:2695]
frames=[dfe1,dfe2]
dfe3=pd.concat(frames)
dfe3=dfe3.sample(frac=1)
dfe3
```

```
#Reseta os indices
dfe3 = dfe3.reset_index()

#Excluir os index anteriores
dfe3 = dfe3.drop(columns=['index'],axis=1)
dfe3
```

# **Analise dos dados**

```
#Verifica a quantidade de pessoas encaminhas para a UTI
dfe3['icu'].value_counts()
```

```
#Separa o conjunto em features (X) e targets (Y)
X = dfe3.iloc[:,:-1]
Y = dfe3.iloc[:,-1]
X.shape, Y.shape
```

```
#A função (train_test_split) os dados de treinamento e teste:  
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```
#Apresentação das dimensões dos conjuntos criados
print('\n', X_train.shape) 
print('\n', Y_train.shape) 
print('\n', X_test.shape) 
print('\n', Y_test.shape)
print('\n')

#Balanceamento de dados
#Treinamento
print(' {0} : ({1}%)'.format(1, (sum(Y_train == 1)/len(Y_train))*100))
print(' {0} : ({1}%)'.format(2, (sum(Y_train == 2)/len(Y_train))*100))

#Teste
print(' {0} : ({1}%)'.format(1, (sum(Y_test == 1)/len(Y_test))*100))
print(' {0} : ({1}%)'.format(2, (sum(Y_test == 2)/len(Y_test))*100))
```

```
#Bibliotecas de importantes
import random
import matplotlib.pyplot as plt


#Bibliotecas de metricas
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```
#Classificador KNN
from sklearn.neighbors import KNeighborsClassifier

#avaliação para KNN
realizacoes = 20

#Matriz que vai salvar medidas de cada realização
scoresKNN = np.zeros([realizacoes,4])

for item in range(realizacoes):
  X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=random.randrange(0, 101, 2))
  neigh = KNeighborsClassifier(n_neighbors=3)
  neighb=neigh.fit(X_train,Y_train)
  Y_pred=neighb.predict(X_test)
  scoresKNN[item,0] = accuracy_score(Y_test,Y_pred)
  scoresKNN[item,1] = f1_score(Y_test,Y_pred, average='weighted')
  scoresKNN[item,2] = recall_score(Y_test,Y_pred, average='micro')
  scoresKNN[item,3] = precision_score(Y_test,Y_pred, average='weighted')

  print('Score {0} : F1 {1} : Recall {2} : Precision {3}'.format(scoresKNN[item,0], scoresKNN[item,1], scoresKNN[item,2], scoresKNN[item,3]))

print('Métricas')
metricsKNN = pd.DataFrame(scoresKNN, columns=['Accuracy', 'F1', 'Recall', 'Precision'])
print(metricsKNN.mean()*100)
confusion_matrix(Y_test,Y_pred)
```

```
#Classificador NB
from sklearn.naive_bayes import GaussianNB

#avaliação para NB
realizacoes = 20

#Matriz que vai salvar medidas de cada realização
scoresNB = np.zeros([realizacoes,4])

for item in range(realizacoes):
  X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=random.randrange(0, 101, 2))
  gnb = GaussianNB()
  Y_pred = gnb.fit(X_train, Y_train).predict(X_test)
  scoresNB[item,0] = accuracy_score(Y_test,Y_pred)
  scoresNB[item,1] = f1_score(Y_test,Y_pred, average='weighted')
  scoresNB[item,2] = recall_score(Y_test,Y_pred, average='micro')
  scoresNB[item,3] = precision_score(Y_test,Y_pred, average='weighted')

  print('Score {0} : F1 {1} : Recall {2} : Precision {3}'.format(scoresNB[item,0], scoresNB[item,1], scoresNB[item,2], scoresNB[item,3]))

print('Métricas')
metricsNB = pd.DataFrame(scoresNB, columns=['Accuracy', 'F1', 'Recall', 'Precision'])
print(metricsNB.mean()*100)
confusion_matrix(Y_test,Y_pred)
```

```
#Classificador LR
from sklearn.linear_model import LogisticRegression

from sklearn.preprocessing import MinMaxScaler

#avaliação para LR
realizacoes = 20

#Matriz que vai salvar medidas de cada realização
scoresLR = np.zeros([realizacoes,4])

for item in range(realizacoes):
  X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=random.randrange(0, 101, 2))
  scaler = MinMaxScaler()
  X_train=scaler.fit_transform(X_train)
  X_test=scaler.transform(X_test)
  clf = LogisticRegression(random_state=0).fit(X_train, Y_train)
  Y_pred = clf.predict(X_test)
  scoresLR[item,0] = accuracy_score(Y_test,Y_pred)
  scoresLR[item,1] = f1_score(Y_test,Y_pred, average='weighted')
  scoresLR[item,2] = recall_score(Y_test,Y_pred, average='micro')
  scoresLR[item,3] = precision_score(Y_test,Y_pred, average='weighted')

  print('Score {0} : F1 {1} : Recall {2} : Precision {3}'.format(scoresLR[item,0], scoresLR[item,1], scoresLR[item,2], scoresLR[item,3]))

print('Métricas')
metricsLR = pd.DataFrame(scoresLR, columns=['Accuracy', 'F1', 'Recall', 'Precision'])
print(metricsLR.mean()*100)
confusion_matrix(Y_test,Y_pred)
```
