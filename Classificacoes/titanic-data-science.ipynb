# **Titanic Data Science Solutions**

```
#libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```
#Load
df = pd.read_csv('https://raw.githubusercontent.com/wellingtondantas/Data-Science/master/Datasets/titanic.csv')
df.head()
```

```
#Number of instances
print(df.shape)

#columns in Data Frame
print(df.columns)

#columns null
df.isnull().sum()
```

```
#Values null
df[df.isnull().any(axis=1)]
```

```
#Drop values null
df.dropna(inplace=True)
```

```
#New Data Frame
ndf = pd.DataFrame(df, columns=['Age','Fare','Sex','sibsp','Parch','Pclass','Embarked','2urvived'] )
ndf
```

```
#Change columns
ndf.columns = ['Age','Fare','Sex','Family','Dependents','Class','Embarked','Survived']
ndf.head()
```

```
#Create age range
conditions = [(ndf['Age'] <= 10), (ndf['Age'] <= 30), (ndf['Age'] > 30)]
choices = [0,1,2]
ndf['Age Range'] = np.select(conditions, choices, default='black')
ndf['Age Range'] = ndf['Age Range'].astype('float64')
```

```
#Create fare range
conditions = [(ndf['Fare'] <= 100), (ndf['Fare'] > 100)]
choices = [0,1]
ndf['Fare Range'] = np.select(conditions, choices, default='black')
ndf['Fare Range'] = ndf['Fare Range'].astype('float64')
```

```
#Show dataframe
ndf.head()
```

```
# Transformation data
from sklearn.preprocessing import MinMaxScaler
```

```
#Norm
adjusted_variables = MinMaxScaler()
raw_variables = ['Fare Range', 'Sex','Class', 'Age Range']
X = adjusted_variables.fit_transform(ndf[raw_variables])
Y = output_variables = ndf['Survived']
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20,random_state=42) 

print(f'Train: {len(x_train)}')
print(f'Test: {len(x_test)}')
```

```
#KNN
from sklearn.neighbors import KNeighborsClassifier 

clf_KNN = KNeighborsClassifier(n_neighbors=5)

 
clf_KNN.fit(x_train, y_train)


KNN = clf_KNN.predict(x_test)
```

```
#DecisionTree (DT)
from sklearn.tree import DecisionTreeClassifier

clf_tree = DecisionTreeClassifier(random_state=1)


clf_tree.fit(x_train, y_train)


DT = clf_tree.predict(x_test)
```

```
#RandomForest (RF)
from sklearn.ensemble import RandomForestClassifier

clf_forest = RandomForestClassifier(max_depth=5, random_state=1)

clf_forest.fit(x_train, y_train)

RF = clf_forest.predict(x_test)
```

```
#SVM
from sklearn.svm import SVC

clf_svm=SVC(gamma='auto',random_state=1)

clf_svm.fit(x_train, y_train)

SVM = clf_svm.predict(x_test)
```

```
#MLP
from sklearn.neural_network import MLPClassifier 

clf_mlp = MLPClassifier(solver='lbfgs', alpha=1e-2, hidden_layer_sizes=(20, 20), random_state=1)

# Treinando o modelo
clf_mlp.fit(x_train, y_train)

# Testando o modelo
MLP = clf_mlp.predict(x_test)
```

```
from sklearn.metrics import accuracy_score

print(f' KNN: {accuracy_score(y_test, KNN)*100}')
print(f' DT: {accuracy_score(y_test, DT)*100}')
print(f' RF: {accuracy_score(y_test, RF)*100}')
print(f' SVM: {accuracy_score(y_test, SVM)*100}')
print(f' MLP: {accuracy_score(y_test, MLP)*100}')
```
