#Analisando a Correlação Entre Ibovespa e Dolar aos longo de Determinado Tempo




```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import seaborn as sns

#!pip install yfinance --upgrade --no-cache-dir
import yfinance as yf
yf.pdr_override()
```

```
cotacoes = ["^BVSP", "USDBRL=X"]
carteira = web.get_data_yahoo(cotacoes)["Close"]
```

```
carteira
```

```
carteira.plot(subplots=True, figsize=(18,5))
```

```
carteira = web.get_data_yahoo(cotacoes, start="2007-01-01")["Close"]
```

```
carteira.plot(subplots=True, figsize=(18,5))
```

```
carteira = carteira.dropna()
carteira
```

```
carteira.columns = ["DOLAR", "BoVESPA"]
sns.set()
carteira.plot(subplots=True, figsize=(18,5))
```

```
sns.heatmap(carteira.corr(), annot=True)
```

```
carteira["DOLAR"].rolling(252).corr(carteira["BoVESPA"]).plot(figsize=(18,5))
```

```
carteira["IBOV_DOLARIZADO"] = (carteira["BoVESPA"]/ carteira["DOLAR"])
```

```
carteira.plot(subplots=True, figsize=(18,7))
```
