```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web

!pip install yfinance --upgrade --no-cache-dir
import yfinance as yf
yf.pdr_override()
```

```
ibovespa = web.get_data_yahoo('^BVSP')
```

```
ibovespa.head()
```

```
ibovespa.tail()
```

```
ibovespa["Close"].plot(figsize=(18,5), label="Histórico do Ibovespa")
ibovespa["Close"].rolling(21).mean().plot(label="MM20")
ibovespa["Close"].rolling(200).mean().plot(label="MM200")
plt.legend()
```

```
ibovespa_fatiado = ibovespa[ibovespa.index.year == 2018]
ibovespa_fatiado["Close"].plot(figsize=(18,5), label="Histórico do Ibovespa")
ibovespa_fatiado["Close"].rolling(21).mean().plot(label="MM20")
ibovespa_fatiado["Close"].rolling(200).mean().plot(label="MM200")
plt.legend()
```

```
ibovespa_fatiado = ibovespa[(ibovespa.index.year >= 2008) & (ibovespa.index.year <= 2015)]
ibovespa_fatiado["Close"].plot(figsize=(18,5), label="Histórico do Ibovespa")
ibovespa_fatiado["Close"].rolling(21).mean().plot(label="MM20")
ibovespa_fatiado["Close"].rolling(200).mean().plot(label="MM200")
plt.legend()
```

```
ibovespa_fatiado = ibovespa[ibovespa.index.month == 12]
ibovespa_fatiado.tail(50)
```

```
ibovespa_fatiado = ibovespa[ibovespa.index.day == 14]
ibovespa_fatiado.tail(50)
```

```
ibovespa = web.get_data_yahoo('VALE3.SA', start='2010-05-03', end='2013-03-01')
```

```
ibovespa["Close"].plot(figsize=(18,5), label="VALE3.SA")
ibovespa["Close"].rolling(7).mean().plot(label="MM7")
ibovespa["Close"].rolling(21).mean().plot(label="MM21")
ibovespa["Close"].rolling(200).mean().plot(label="MM200")
plt.legend()
```
